```python
from transformers import GPT2TokenizerFast
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

# remove tweets that are too long
train_data = train_data.dropna()
train_data['n_tokens'] = train_data.text.apply(lambda x: len(tokenizer.encode(x)))
train_data = train_data[train_data.n_tokens<2000].tail(1_000)
train_data
```

```python
from transformers import GPT2TokenizerFast
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

# removing long tweets
test_data['n_tokens'] = test_data.text.apply(lambda x: len(tokenizer.encode(x)))
test_data = test_data[test_data.n_tokens<2000].tail(1_000)
test_data
```

```python
%%time
import time
from openai.embeddings_utils import get_embedding
i = 0
df = train_data.head(0)
while i <1000:
  sub = train_data.iloc[i:i+5, :]
  sub['babbage_similarity'] = sub.text.apply(lambda x: get_embedding(x, engine='text-similarity-babbage-001'))
  i = i+5
  df = pd.concat([df, sub])
  time.sleep(10)
  
df.head()
df.to_csv('embeddings.csv')
```

```python
data_path = "/content/embeddings.csv"
df = pd.read_csv(data_path)
df["babbage_similarity"] = df.babbage_similarity.apply(eval).apply(np.array)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    list(df.babbage_similarity.values), df.label, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
probas = clf.predict_proba(X_test)

report = classification_report(y_test, preds)
print(report)
```
